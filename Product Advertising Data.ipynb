```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/Users/lisa/Desktop/作品集/Product Advertising Data in kaggle/Advertising_Data.csv')
```

```python
#各渠道 CPA
ad_channel = ['TV','Billboards','Google_Ads','Social_Media','Influencer_Marketing','Affiliate_Marketing']
channel_CPA = {}

for channel in ad_channel:
    per_channel_spent = df[channel].sum()
    total_product_sold = df['Product_Sold'].sum()
    CPA = per_channel_spent / total_product_sold
    channel_CPA[channel] = CPA
    
avg_cpa_series = pd.Series(channel_CPA).sort_values(ascending = False)

#各渠道相關係數

corr = df[ad_channel].corrwith(df['Product_Sold']).sort_values(ascending = False)
```

```python
avg_cpa_series.plot(kind = 'barh' , title = 'Average CPA Comparison by Channel' , xlabel = 'Average CPA' , ylabel = 'Ad channel')
```

```python
corr.plot(kind = 'bar' , title = 'Correlation between Channel Spend and Product Sold' , xlabel = 'Ad channel' , ylabel = 'Correlation Coefficient (r)')
```
